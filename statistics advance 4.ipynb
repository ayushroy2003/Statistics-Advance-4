### Q8. What is the margin of error in a confidence interval? How does sample size affect the margin of error? Provide an example of a scenario where a larger sample size would result in a smaller margin of error.

** Margin of Error (MOE):**
The margin of error in a confidence interval measures the extent of the uncertainty or potential error in estimating a population parameter (e.g., mean). It represents the range within which the true population parameter is expected to lie with a certain level of confidence (e.g., 95%).

**Formula:**
$$
\text{Margin of Error} = z \times \frac{\sigma}{\sqrt{n}}
$$

Where:
- \( z \) = z-score corresponding to the confidence level.
- \( \sigma \) = population standard deviation.
- \( n \) = sample size.

**Effect of Sample Size:**
As the sample size increases, the denominator \( \sqrt{n} \) increases, causing the margin of error to decrease. This means larger sample sizes lead to more precise estimates (smaller margin of error).

**Example:**
Imagine you're conducting a survey to estimate the average amount of time people spend on social media daily. If you survey 100 people, your margin of error might be 2 hours. However, if you survey 1,000 people, the margin of error might decrease to 0.6 hours, providing a more precise estimate of the population's behavior.
        


### Q9. Calculate the z-score for a data point with a value of 75, a population mean of 70, and a population standard deviation of 5. Interpret the results.

**Formula:**
$$
z = \frac{X - \mu}{\sigma}
$$

Where:
- \( X = 75 \) (data point)
- \( \mu = 70 \) (mean)
- \( \sigma = 5 \) (standard deviation)

**Calculation:**
$$
z = \frac{75 - 70}{5} = \frac{5}{5} = 1
$$

**Interpretation:**
A z-score of 1 indicates that the data point (75) is 1 standard deviation above the mean. This means that this value is slightly higher than average, but within the normal range of variation.
        


### Q10. In a study of the effectiveness of a new weight loss drug, a sample of 50 participants lost an average of 6 pounds with a standard deviation of 2.5 pounds. Conduct a hypothesis test to determine if the drug is significantly effective at a 95% confidence level using a t-test.

**Step 1: Define Hypotheses**
- Null Hypothesis (\( H_0 \)): The drug is not effective (\( \mu = 0 \)).
- Alternative Hypothesis (\( H_1 \)): The drug is effective (\( \mu \neq 0 \)).

**Step 2: Calculate the t-statistic**
$$
t = \frac{\bar{X} - \mu_0}{\frac{s}{\sqrt{n}}}
$$

Where:
- \( \bar{X} = 6 \) (sample mean)
- \( \mu_0 = 0 \) (hypothesized mean)
- \( s = 2.5 \) (sample standard deviation)
- \( n = 50 \) (sample size)

**Calculation:**
$$
t = \frac{6 - 0}{\frac{2.5}{\sqrt{50}}} = \frac{6}{\frac{2.5}{7.071}} = \frac{6}{0.3536} \approx 16.97
$$

**Step 3: Determine the critical t-value**
For a two-tailed test at a 95% confidence level and \( df = 50 - 1 = 49 \), the critical t-value is approximately ±2.009 (from t-distribution tables).

**Step 4: Compare t-statistic with critical t-value**
Since \( t = 16.97 \) is much greater than 2.009, we reject the null hypothesis.

**Conclusion:**
The t-test shows that the drug is significantly effective in weight loss.
        

Q11. In a survey of 500 people, 65% reported being satisfied with their current job. Calculate the 95%
confidence interval for the true proportion of people who are satisfied with their job.

Ans -


### Confidence Interval Formula

The confidence interval for a proportion is given by:

$$\text{Confidence Interval} = \hat{p} \pm Z \left( \frac{\sqrt{\hat{p}(1 - \hat{p})}}{n} \right) $$

Where:
- \(\hat{p}\) represents the sample proportion (in this case, 65% or 0.65).
- \(Z\) corresponds to the Z-score for the desired confidence level (for a 95% confidence interval, \(Z = 1.96\)).
- \(n\) is the sample size (500 in this example).

$$\hat{p} = \frac{65}{100} = 0.65$$

### Calculating standard error 

$$ SE = \sqrt{\frac{0.65(1 - 0.65)}{500}}  = \sqrt{\frac{0.65 \times 0.35}{500}}  = \sqrt{\frac{0.2275}{500}}  = \sqrt{\frac{0.000455}{500}} = \sqrt{0.000455} \approx 0.0213 $$

### Calculating margin of error

$$ME = Z \cdot SE = 1.96 \cdot \sqrt{0.0213} \approx 0.0418$$


### Calculating confidence interval 

Lower Limit: $\hat{p} - ME = 0.65 - 0.0418 = 0.6082$


Upper Limit: $\hat{p} + ME = 0.65 + 0.0418 = 0.6918$

The 95% confidence interval for the true proportion of people who are satisfied with their job is approximately 
0.6082
0.6082 to 
0.6918
0.6918, or in percentage terms, 60.82% to 69.18%.

Q12 A researcher is testing the effectiveness of two different teaching methods on student performance. Sample A has a mean score of 85 with a standard deviation of 6, while sample B has a mean score of 82 with a standard deviation of 5. We will conduct a hypothesis test to determine if the two teaching methods have a significant difference in student performance using a t-test with a significance level of 0.01.

Ans - 
## Step 1: State the Hypotheses
- **Null Hypothesis ($H_0$)**: There is no significant difference in the mean scores between the two teaching methods. $ \mu_A = \mu_B $
- **Alternative Hypothesis ($H_1$)**: There is a significant difference in the mean scores between the two teaching methods. $ \mu_A \neq \mu_B $

## Step 2: Gather the Data
- Mean of Sample A ($\bar{X}_A$) = 85
- Standard deviation of Sample A ($S_A$) = 6
- Size of Sample A ($n_A$) = Not provided, we assume $n_A = n_B$ if not given
- Mean of Sample B ($\bar{X}_B$) = 82
- Standard deviation of Sample B ($S_B$) = 5
- Size of Sample B ($n_B$) = Not provided, we assume $n_B = n_A$ if not given

## Step 3: Determine the Significance Level
The significance level ($\alpha$) is 0.01.

## Step 4: Calculate the Test Statistic
The formula for the t-test statistic for independent samples is:
$$ t = \frac{\bar{X}_A - \bar{X}_B}{\sqrt{\frac{S_A^2}{n_A} + \frac{S_B^2}{n_B}}} $$
Given that the sample sizes are not provided, we'll assume they are equal ($n_A = n_B = n$) for simplicity.
$$ t = \frac{85 - 82}{\sqrt{\frac{6^2}{n} + \frac{5^2}{n}}} = \frac{3}{\sqrt{\frac{36}{n} + \frac{25}{n}}} = \frac{3}{\sqrt{\frac{61}{n}}} = \frac{3\sqrt{n}}{\sqrt{61}} $$

## Step 5: Determine the Critical t-value
To find the critical t-value, we need the degrees of freedom. Since we assumed equal sample sizes:
$$ df = n_A + n_B - 2 = 2n - 2 $$
For $\alpha = 0.01$ in a two-tailed test, the critical t-value can be found using a t-distribution table.

## Step 6: Compare the Test Statistic to the Critical Value
- If $ |t| > t_{critical} $, we reject the null hypothesis.
- If $ |t| \leq t_{critical} $, we fail to reject the null hypothesis.

## Step 7: Draw a Conclusion
If the calculated t-value exceeds the critical value for the given degrees of freedom, we conclude that there is a significant difference between the two teaching methods at the 0.01 significance level. Otherwise, we conclude that there is no significant difference.

**Note**: The exact t-value or critical value cannot be computed without the sample size ($n$). If the sample size is provided, the calculation can be completed.


### Q13. A population has a mean of 60 and a standard deviation of 8. A sample of 50 observations has a mean of 65. Calculate the 90% confidence interval for the true population mean.

**Formula:**
$$
\text{CI} = \bar{X} \pm z \times \frac{\sigma}{\sqrt{n}}
$$

Where:
- \( \bar{X} = 65 \)
- \( \sigma = 8 \)
- \( n = 50 \)
- \( z = 1.645 \) (z-score for 90% confidence level)

**Calculation:**
$$
\text{Margin of Error} = 1.645 \times \frac{8}{\sqrt{50}} = 1.645 \times 1.131 = 1.86
$$

$$
\text{CI} = 65 \pm 1.86 = (63.14, 66.86)
$$

**Conclusion:**
The 90% confidence interval for the true population mean is between 63.14 and 66.86.
        


### Q14. In a study of the effects of caffeine on reaction time, a sample of 30 participants had an average reaction time of 0.25 seconds with a standard deviation of 0.05 seconds. Conduct a hypothesis test to determine if the caffeine has a significant effect on reaction time at a 90% confidence level using a t-test.

**Step 1: Define Hypotheses**
- Null Hypothesis (\( H_0 \)): Caffeine has no effect on reaction time (\( \mu = 0 \)).
- Alternative Hypothesis (\( H_1 \)): Caffeine affects reaction time (\( \mu \neq 0 \)).

**Step 2: Calculate the t-statistic**
$$
t = \frac{\bar{X} - \mu_0}{\frac{s}{\sqrt{n}}}
$$

Where:
- \( \bar{X} = 0.25 \)
- \( \mu_0 = 0 \)
- \( s = 0.05 \)
- \( n = 30 \)

**Calculation:**
$$
t = \frac{0.25 - 0}{\frac{0.05}{\sqrt{30}}} = \frac{0.25}{0.0091} \approx 27.47
$$

**Step 3: Determine the critical t-value**
For a two-tailed test at a 90% confidence level and \( df = 30 - 1 = 29 \), the critical t-value is approximately ±1.699.

**Step 4: Compare t-statistic with critical t-value**
Since \( t = 27.47 \) is much greater than 1.699, we reject the null hypothesis.

**Conclusion:**
The t-test indicates that caffeine significantly affects reaction time.
